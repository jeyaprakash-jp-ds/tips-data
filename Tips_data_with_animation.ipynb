```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import plotly.express as px
```

```python
df=px.data.tips()
df
```

```python
px.histogram(df,x="total_bill",nbins=20,color="sex",title="sex_wise_distribution")
```

```python
px.pie(df,names="day",values="tip")
```
